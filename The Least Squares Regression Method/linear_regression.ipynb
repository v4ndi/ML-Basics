```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

x_mean = x.mean()
y_mean = y.mean()

plt.plot(x, y, 'o')
plt.title("Initial data")
plt.xlabel('X') 
plt.ylabel('Y') 
plt.show()
```

```python
x_line = np.linspace(0, max(x), len(x))
y_line = [y_mean] * len(x_line)

y_line1 = np.linspace(0, max(y), len(x))
x_line1 = [x_mean] * len(x_line)

plt.plot(x_line,y_line, 'r')
plt.plot(x_line1,y_line1, 'b')
plt.plot(x, y, 'o')

plt.title("Mean lines")

plt.xlabel('X') 
plt.ylabel('Y') 

plt.show()
```

```python
x_ = x - x_mean
y_ = y - y_mean

## Find square of x_
x_sqr = x_ * x_

## Find multiply vector x_ by y_
c = x_ * y_

## last constants of equation y = b0 + b1*x
b1 = c.sum() / x_sqr.sum()
b0 = y_mean - b1 * x_mean
```

```python

regression_x = np.array([0, max(x)])
regression_y = b0 + b1*regression_x

plt.plot(x, y, 'o')
plt.plot(regression_x,regression_y)


plt.title("Linear regression")

plt.xlabel('X') 
plt.ylabel('Y') 

plt.show()
```
